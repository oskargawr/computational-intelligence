```python
import nltk

nltk.download('all')
```

```python
article_path = 'article.txt'

with open(article_path, 'r') as file:
    article = file.read()
article
```

```python
import re

article = re.sub(r"[^\w\s]", "", article)
# also delete newlines
article = re.sub(r"\n", "", article)
```

```python
tokens = nltk.word_tokenize(article.lower())
tokens
```

```python
len(tokens)
```

```python
# delete stopwords
from nltk.corpus import stopwords

filtered_tokes = [word for word in tokens if word not in stopwords.words('english')]
```

```python
len(tokens)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(filtered_tokes)
bag_of_words = X.toarray()
bag_of_words
```

```python
vectorizer.get_feature_names_out()
```

```python
import spacy
```

```python
spacy.cli.download('en_core_web_sm')
```

```python
nlp = spacy.load("en_core_web_sm")
doc = nlp(article)

lemmatized_words = [token.lemma_ for token in doc]
lemmatized_words
```

```python
len(lemmatized_words)
```

```python
from collections import Counter

word_freq = Counter(lemmatized_words)
word_freq
```

```python
import matplotlib.pyplot as plt

most_common_words = word_freq.most_common(10)
words, counts = zip(*most_common_words)
plt.figure(figsize=(10, 5))
plt.bar(words, counts)
plt.xlabel("Słowa")
plt.ylabel("Liczba wystąpień")
plt.title("10 najczęściej występujących słów")
plt.show()
```

```python
from wordcloud import WordCloud

wordcloud = WordCloud(width=800, height=400, background_color="white").generate(
    " ".join(lemmatized_words)
)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
